```
import pickle 

# import with pickle
filename = 'gdrive/My Drive/MosquitoNet/mosquito_dataset.pkl'
infile = open(filename,'rb')
dataset = pickle.load(infile)
infile.close()
```

```
label_values = [example["label"] for example in dataset]
# display(label_values)

classes = ['Unfed', 'Blood Fed', 'Half Gravid', 'Gravid']
class_nums = [10, 7, 0, 3]
```

```
fig, ax = plt.subplots(3, 4)

for n,num in enumerate(class_nums):
  example = dataset[num]

  ax[0, n].imshow(example["headlegs"])
  ax[0, n].set_xticks([])
  ax[0, n].set_yticks([])
  ax[0, n].set_title(example["label"])

  ax[1, n].imshow(example["wings"])
  ax[1, n].set_xticks([])
  ax[1, n].set_yticks([])

  ax[2, n].imshow(example["palps"])
  ax[2, n].set_xticks([])
  ax[2, n].set_yticks([])

  if n==0:
    ax[0,n].set_ylabel("Head-Legs")
    ax[1,n].set_ylabel("Wings")
    ax[2,n].set_ylabel("Palps")

plt.savefig('gdrive/My Drive/MosquitoNet/figures/dataset_examples.png')
plt.show()
```
