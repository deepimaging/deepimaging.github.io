# Physical Layers
One of the unique aspects of this class is our discussion of *physical layers*. This is a relatively new concept that is gaining popularity within the machine learning community.

Physical layers, or *learned sensing* is the idea of modifying how data is collected to better suit automated analysis.

In conventional sensing the data is captured based, then at a later point it is processed by a machine learning algorithm. The two are completely seperate processes.

Learned sensing is the idea of combining the two, such that a machine learning algorithm influences data collection. This has the advantage of either reducing capture overhead, or improving performance (or both!). But has the disadvantage of requiring a more sophisticated setup to capture training data.

## Examples of Learned Sensing in Microscopy

Learned sensing is all about **sensing**, and nothing more. It is about modifying, or simulating a modification, do how data is being captured. It is not about new architectures, or clever pre-processing schemes

## Exercise 1A - Create a classification model

To speed things I have provided code to generate a simple classification model

````
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPool2D, Dense, Flatten, BatchNormalization

def create_model(input_size=(28,28,1)):
    input_layer = Input(input_size)
    x = input_layer
    x= BatchNormalization()(x)
    num_filters = 8
    for i in range(3):
        x = Conv2D(num_filters*(2**i), 3, activation='relu', padding='same')(x)
        x = Conv2D(num_filters*(2**i), 3, activation='relu')(x)
        if i == 0:
            x = BatchNormalization()(x)
        x = MaxPool2D()(x)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dense(10, activation='softmax')(x)
    return tf.keras.Model(inputs=input_layer, outputs=x)
````
